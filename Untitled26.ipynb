```python
# 5. Customer Segmentation Based on Purchasing Patterns
To segment customers, we can use recency, frequency, and monetary (RFM) analysis. Here we will create segments based on the recency of purchases and total spend.
RFM Segmentation (Recency, Frequency, Monetary)
```

```python
print(data_clean.columns)
```

```python
# Calculate recency: Days since the last purchase
last_purchase_date = data_clean.groupby('Customer ID')['InvoiceDate'].max()
recency = (data_clean['InvoiceDate'].max() - last_purchase_date).dt.days

# Frequency: Number of purchases per customer
frequency = data_clean.groupby('Customer ID')['Invoice'].nunique()

# Monetary: Total spend per customer
monetary = data_clean.groupby('Customer ID')['Total'].sum()

# Combine into a DataFrame
rfm = pd.DataFrame({'Recency': recency, 'Frequency': frequency, 'Monetary': monetary})

# Create segments based on quantiles (with handling for duplicates)
rfm['Recency_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])  # Lower score means more recent

# Get the unique quantiles for Frequency and adjust labels accordingly
n_freq_bins = pd.qcut(rfm['Frequency'], 4, duplicates='drop').nunique()
rfm['Frequency_Score'] = pd.qcut(rfm['Frequency'], 4, labels=range(1, n_freq_bins + 1), duplicates='drop')

# Get the unique quantiles for Monetary and adjust labels accordingly
n_monetary_bins = pd.qcut(rfm['Monetary'], 4, duplicates='drop').nunique()
rfm['Monetary_Score'] = pd.qcut(rfm['Monetary'], 4, labels=range(1, n_monetary_bins + 1), duplicates='drop')

# Total RFM score (just a sum of individual scores)
rfm['RFM_Score'] = rfm['Recency_Score'].astype(int) + rfm['Frequency_Score'].astype(int) + rfm['Monetary_Score'].astype(int)

# Segment customers based on RFM score
rfm['Segment'] = pd.cut(rfm['RFM_Score'], bins=[0, 3, 5, 7, 9], labels=['Low', 'Mid', 'High', 'Very High'])

# Show the RFM segments
print(rfm['Segment'].value_counts())
```

```python
# Plot the distribution of customer segments
rfm['Segment'].value_counts().plot(kind='bar', color='lightseagreen', figsize=(8,6))
plt.title('Customer Segmentation Based on RFM Score')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.show()
```
